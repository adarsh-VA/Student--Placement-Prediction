```python
import pandas as pd
import numpy as np
import random as r
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import ExtraTreeClassifier

from sklearn.metrics import accuracy_score as acc
```

```python
data = pd.read_csv("dat2.csv")
```

```python
data.columns =['gender','English','Java','Python','Machine Learning']
```

```python
data
```

```python
j = []
for i in range(664):
    j.append(r.randint(0,10))

data["marks"] = j
data
```

```python
g = []
for i in range(len(data)):
    g.append(data.iloc[i].values[1:].sum())

data["sum"] = g
data
```

```python
data.describe()
```

```python
target_uni = ["Data scientist","full-stack dev","tech-support","non-tech support","finance"]
s = []
sc=0
for i in range(len(data)):
    sc=data["sum"].iloc[i]
    if sc >14:
        s.append(0)
    elif sc >10 and sc<15:
        s.append(1)
    elif sc >7 and sc<11:
        s.append(2)
    elif sc >4 and sc<8:
        s.append(3)
    else:
        s.append(4)
s.count(4)
```

```python
s.count(4)
```

```python
data["target"] = s
data
```

```python
data.drop(columns=["sum"],inplace=True)
```

```python
data
```

```python
X = data.iloc[:, 0:6].values
y = data.iloc[:, 6].values
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
rfr = RandomForestClassifier()
rfr.fit(X_train,y_train)
```

```python
y_rfr = rfr.predict(X_test)
acc(y_rfr,y_test)
```

```python
etc = ExtraTreeClassifier()
etc.fit(X_train,y_train)
```

```python
y_etc = etc.predict(X_test)
acc(y_etc,y_test)
```
